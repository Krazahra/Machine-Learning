```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/housepricedata.csv')
df.head()
```

```python
df.info()
```

```python
# Set the target column and feature columns
target_column = 'AboveMedianPrice'
X = df.drop(columns=[target_column])
y = df[target_column]

# Normalize the feature columns using MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the df into 70% training and 30% validation and test
X_train, X_val_test, y_train, y_val_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Further split the validation and test df into 50% validation and 50% test
X_val, X_test, y_val, y_test = train_test_split(X_val_test, y_val_test, test_size=0.5, random_state=42)
```

```python
# Create the neural network model
model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train,
                    validation_data=(X_val, y_val),
                    batch_size=32,
                    epochs=300)

# Evaluate the model on the test df
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```
