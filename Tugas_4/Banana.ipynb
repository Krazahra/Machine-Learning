```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import pandas as pd

# Statistik deskriptif untuk fitur numerik
numeric_stats = df.describe()

# Statistik deskriptif untuk fitur kategori
categorical_stats = df.describe(include=['object'])

# Menampilkan statistik deskriptif
print("Statistik Deskriptif untuk Fitur Numerik:")
print(numeric_stats)

print("\nStatistik Deskriptif untuk Fitur Kategori:")
print(categorical_stats)
```

```python
# Mengambil hanya fitur numerik
numeric_features = df.select_dtypes(include=['int64', 'float64'])

# Menampilkan boxplot untuk setiap fitur numerik
plt.figure(figsize=(10, 6))
numeric_features.boxplot()
plt.title('Boxplot untuk Fitur Numerik')
plt.ylabel('Nilai')
plt.xticks(rotation=45)  # Putar label sumbu x untuk lebih mudah dibaca jika diperlukan
plt.show()
```

```python
# Mengambil hanya fitur numerik
numeric_features = df.select_dtypes(include=['int64', 'float64'])

# Membuat scatter plot untuk semua pasangan fitur numerik
sns.pairplot(data=numeric_features)
plt.suptitle('Scatter Plot untuk Fitur Numerik', y=1.02)
plt.show()
```

```python
# Mengambil hanya fitur numerik
numeric_features = df.select_dtypes(include=['int64', 'float64'])

# Menghitung korelasi antara fitur numerik
correlation_matrix = numeric_features.corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Heatmap Korelasi antara Fitur Numerik')
plt.show()
```
